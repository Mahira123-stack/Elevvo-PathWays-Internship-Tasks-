**Clean the Data**

```python
df.isnull().sum()
```

```python
df.drop(columns=['Cabin'], inplace=True)
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

**Convert Data Types**

```python
df.dtypes
```

```python
df['Sex'] = df['Sex'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
```

```python
df.dtypes
```

**Generate Group-Based Insights**

```python
# Survival by Gender

df.groupby('Sex')['Survived'].mean()
```

```python
# Survival by Class

df.groupby('Pclass')['Survived'].mean()
```

```python
df.pivot_table(index='Sex', columns='Pclass', values='Survived')
```

**Visualizing Key Patterns**

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Fare vs Age (colored by Survival)')
plt.show()
```

```python
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Not Survived', 'Survived'], colors=['lightcoral', 'lightgreen'])
plt.title('Overall Survival Rate')
plt.ylabel('')
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
plt.show()
```

```python
# Drop text-based columns
numeric_df = df.drop(columns=['Name', 'Ticket'])

# Create correlation matrix
corr = numeric_df.corr(numeric_only=True)

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Survival by Age (Histogram)
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', kde=True)
plt.title('Survival by Age')
plt.show()
```

```python
sns.catplot(x='Pclass', hue='Survived', col='Sex', kind='count', data=df)
```

```python
# Create pivot table
pivot = df.pivot_table(index='Sex', columns='Pclass', values='Survived')

# Plot heatmap
sns.heatmap(pivot, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Survival Rate by Gender and Class')
plt.show()
```
