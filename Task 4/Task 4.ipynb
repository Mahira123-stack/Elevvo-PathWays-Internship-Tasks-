### **Drop duplicates and handle missing values**

```python
df.isnull().sum()
```

```python
print("\nMissing Values (%)\n", (df.isnull().mean() * 100).sort_values(ascending=False).head(10))
```

```python
print("\nDuplicates:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
threshold = 0.7
df = df.loc[:, df.isnull().mean() < threshold]
```

### **Fix Inconsistent Formatting**

```python
# Strip spaces, make consistent case
df['Q3'] = df['Q3'].str.strip().str.title()  # Q3 in 2020 is "Country"

# Example: Replace variations
df['Q3'] = df['Q3'].replace({
    'United States of America': 'USA',
    'United Kingdom of Great Britain and Northern Ireland': 'UK'
})
```

### **Applying label encoding**

```python
from sklearn.preprocessing import LabelEncoder

# Make a copy so we don't overwrite original
df_encoded = df.copy()

label_encoders = {}
for col in df_encoded.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    label_encoders[col] = le  # store encoder for future decoding
```

### **Extract meaningful insights**

**Top 10 Countries by Participation**

```python
plt.figure(figsize=(10,5))
df['Q3'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Countries by Survey Participation")
plt.ylabel("Number of Respondents")
plt.xlabel("Country")
plt.show()
```

```python
# Education Level Distribution

plt.figure(figsize=(8,4))
sns.countplot(y='Q4', data=df, order=df['Q4'].value_counts().index)
plt.title("Education Level of Respondents")
plt.show()
```

```python
# Preferred Programming Languages

lang_cols = [col for col in df.columns if 'Q7' in col]  # Q7 = Programming Languages
lang_counts = df[lang_cols].apply(lambda x: x.value_counts()).sum(axis=1).sort_values(ascending=False)

lang_counts.head(10).plot(kind='bar', color='orange')
plt.title("Most Popular Programming Languages")
plt.ylabel("Number of Mentions")
plt.show()
```

```python
# Tools Preference by Job Role

job_tool = df.groupby('Q5')['Q7_Part_1'].value_counts().unstack().fillna(0)  # Q5 = Role, Q7_Part_1 = Python
plt.figure(figsize=(12,6))
sns.heatmap(job_tool, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title("Python Usage by Job Role")
plt.show()
```

```python
# Cloud Platform Usage

cloud_cols = [col for col in df.columns if 'Q27' in col]  # Cloud platforms question
cloud_counts = df[cloud_cols].apply(lambda x: x.value_counts()).sum(axis=1).sort_values(ascending=False)

cloud_counts.head(5).plot(kind='bar', color='teal')
plt.title("Most Popular Cloud Platforms")
plt.ylabel("Number of Mentions")
plt.show()
```

### **Summary dashboard showing top 5 insights**

```python
fig, axes = plt.subplots(3, 2, figsize=(14, 12))
fig.suptitle("Kaggle Survey — Top 5 Insights", fontsize=18, fontweight='bold')
axes = axes.flatten()

# 1️⃣ Top 5 Countries
df['Q3'].value_counts().head(5).plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title("Top 5 Countries")
axes[0].set_ylabel("Respondents")

# 2️⃣ Education Level
sns.countplot(y='Q4', data=df, order=df['Q4'].value_counts().index, ax=axes[1], palette='pastel')
axes[1].set_title("Education Level Distribution")

# 3️⃣ Programming Languages
lang_cols = [col for col in df.columns if 'Q7' in col]
lang_counts = df[lang_cols].apply(lambda x: x.value_counts()).sum(axis=1).sort_values(ascending=False)
lang_counts.head(5).plot(kind='bar', ax=axes[2], color='orange')
axes[2].set_title("Top 5 Programming Languages")

# 4️⃣ Salary by Age
sns.boxplot(x='Q1', y='Q24', data=df, ax=axes[3])
axes[3].set_title("Salary Range by Age Group")
axes[3].tick_params(axis='x', rotation=45)

# 5️⃣ Cloud Platforms
cloud_cols = [col for col in df.columns if 'Q27' in col]
cloud_counts = df[cloud_cols].apply(lambda x: x.value_counts()).sum(axis=1).sort_values(ascending=False)
cloud_counts.head(5).plot(kind='bar', ax=axes[4], color='teal')
axes[4].set_title("Top 5 Cloud Platforms")

# Remove last empty subplot (since we only have 5 insights)
fig.delaxes(axes[5])

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
